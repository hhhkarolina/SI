```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import numpy as np
from sklearn.model_selection import train_test_split
from bs4 import BeautifulSoup
import os
from sentence_transformers import SentenceTransformer
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
# from sklearn.neural import network.MLPRegressor
```

# Podział na zbiór uczący i walidacyjny

```python
file = "jester-data-1.xls"
sheet = "jester-data-1-new"

# Wczytanie danych za pomocą pandas
df = pd.read_excel(file, sheet_name = sheet, header = None)

# Wyświetlanie zawartości DataFrame
df
```

```python
df = df.replace(99, np.nan)
df_mean = pd.DataFrame(df.mean())
df_mean = df_mean.drop(0)
df_mean
```

```python
df_std = pd.DataFrame(df.std())
df_std
```

```python
folder = 'jokes'
jokes_files = os.listdir(folder)
jokes = []

for file in jokes_files:
    joke_path = os.path.join(folder, file)
    if os.path.isfile(joke_path):
        with open(joke_path, 'r') as joke_file:
            joke = joke_file.read()
            jokes.append(joke)

jokes.pop(0)
```

```python
start = "<!--begin of joke -->"
end = "<!--end of joke -->"
cut_jokes = []
clean_jokes = []

for joke in jokes:
    start_index = joke.find(start)
    end_index = joke.find(end)

    if start_index != -1 and end_index != -1:
        start_index += len(start)
        cut_fragment = joke[start_index:end_index].strip()
        cut_jokes.append(cut_fragment)

for joke in cut_jokes:
    joke = joke.replace('\n', '')
    soup = BeautifulSoup(joke, "html.parser")
    text = soup.get_text()
    clean_jokes.append(text)
    
clean_jokes
```

```python
model = SentenceTransformer('bert-base-cased')

texts_list = clean_jokes

embeddings = model.encode(texts_list)

print(embeddings.shape)
```

```python
X = embeddings
Y = df_mean.values.ravel()

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size = 0.2)


scaler = StandardScaler()  
scaler.fit(x_train)  
x_train = scaler.transform(x_train)  
x_test = scaler.transform(x_test)  
```

#  Działanie podstawowego modelu MLP

```python
mlp = MLPRegressor(
    learning_rate = 'constant',  
    solver = 'sgd', 
    max_iter = 1, 
    warm_start = True, 
    learning_rate_init = 0.001,
)

val_loss = []
train_loss = []

    
for epoch in range(200):
    # Trenowanie modelu w kolejnej epoce
    mlp.fit(x_train, y_train)
    
    # Obliczanie i zapisywanie wartości błędu treningowego  
    train_loss.append(mlp.loss_)
    
    # Obliczanie i zapisywanie wartości błędu walidacyjnego
    y_test_pred = mlp.predict(x_test)
    val_loss.append(mean_squared_error(y_test, y_test_pred))

# Wizualizacja wartości funkcji kosztu
epochs = range(1, mlp.n_iter_ + 1)

plt.plot(epochs, train_loss, label='Train Loss')
plt.plot(epochs, val_loss, label='Validation Loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Predykcja na zbiorze testowym
y_pred = mlp.predict(x_test)
plt.plot(y_pred, c = 'red')
plt.plot(y_test, c = 'blue')
plt.xticks(range(0, 20, 1))
plt.show()

mse = mean_squared_error(y_test, y_pred)
print("Błąd średniokwadratowy (MSE): {:.2f}".format(mse))
```
